# Modelling

Before fitting our data to the model, we will have to label encode our categorical data, scale our numerical data for better performance.
Label Encoder converts categorical features to numeric values. To scale our numeric values, we will use RobustScaler which is robust to OUTLIERS!
Based on feature correlation with our target variable, we will create some new features. 

Our target is to predict 0 : if employee leaves the organization OR 1 : if employee stays

In this exercise, I will be using Regression Algorithms to predict Number of days after which an employee will leave the organization since joining(Difference).


The condition is:
    if  01 January '18 < number of days < 01 July '18
        then we classify as 1 else 0

I will be testing the performance on a few base models. Based of RMSE metric and model predictions, I will be choosing two model equally weighted for final prediction

```python
train_df = train.copy(deep=True)
test_df = test.copy(deep=True)

train_df.drop(columns =['Emp_ID'], inplace=True)
test_df.drop(columns =['Emp_ID'], inplace=True)
```

Base Models

```python
n_folds = 10

def rmsle_cv(model):
    kf = KFold( n_folds, shuffle=True, random_state=0 ).get_n_splits(X)
    rmse = np.sqrt( -cross_val_score(model,X, y, scoring= 'neg_mean_squared_error', cv =kf) )
    return (rmse)
```

```python
lr = make_pipeline( RobustScaler(),  LinearRegression() )

lasso = make_pipeline( RobustScaler(), Lasso(alpha=0.5, random_state=1))

ENet = make_pipeline( RobustScaler(), ElasticNet(alpha=0.5, l1_ratio=0.5, random_state=0))

KRR = KernelRidge()

dt = DecisionTreeRegressor()

rf = RandomForestRegressor()

GBoost = GradientBoostingRegressor(random_state=0, n_estimators=200)

model_xgb = xgb.XGBRegressor(random_state=0)

model_lgb = lgb.LGBMRegressor(objective='regression', n_estimators=200, reg_lambda=0.5)
```

```python
from pandas.tseries.offsets import DateOffset

sub['Leave_Date_Pred'] = 0
for i in range(len(sub)):
    #print(i)
    n1 = sub['Date_Of_Joining'][i]
    n2 = sub['Difference'][i]

    n3 = (n1 + DateOffset(days=int(n2)))
    sub['Leave_Date_Pred'][i] = n3
#sub
```

```python
startDate = pd.to_datetime('01-01-2018')
endDate= pd.to_datetime('07-01-2018')

sub['Target'] = 0

for j in range(len(sub)):
    if startDate< sub['Leave_Date_Pred'][j] < endDate:
        sub['Target'][j] = 1
    else:
        sub['Target'][j] = 0
sub
```
