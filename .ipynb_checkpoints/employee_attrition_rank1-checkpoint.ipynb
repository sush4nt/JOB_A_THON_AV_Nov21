```python
#Please Note, This is just a solution not my research
import pandas as pd
import numpy as np
from pandas.core.common import SettingWithCopyWarning
import warnings
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

train = pd.read_csv('https://datahack-prod.s3.amazonaws.com/train_file/train_MpHjUjU.csv')
test = pd.read_csv('https://datahack-prod.s3.amazonaws.com/test_file/test_hXY9mYw.csv')

train.City = train.City.str.slice(1)
train.Gender.replace('Female',0,inplace=True)
train.Gender.replace('Male',1,inplace=True)

train.Education_Level.replace('College',0,inplace=True)
train.Education_Level.replace('Bachelor',1,inplace=True)
train.Education_Level.replace('Master',2,inplace=True)
train.Dateofjoining = pd.to_datetime(train.Dateofjoining)
train['MMM-YY'] = pd.to_datetime(train['MMM-YY'])
# age of emp or employ tenure
train['emp_tenure'] = (train['MMM-YY'] - train.Dateofjoining).dt.days
train['Target'] = np.where(train.LastWorkingDate.isnull(),0,1)
train.drop(['LastWorkingDate','Dateofjoining'],axis=1,inplace=True)

# we are doing this because when an employe is about leave an organization we can start showing change in behaviour before churn 
def churn(dataset, c=9): 
  emp_ids = dataset.Emp_ID.unique()
  account_data_list = []
  for i in emp_ids:
    account_data_temp = dataset[dataset.Emp_ID == i]
    if account_data_temp['Target'].sum() != 0:
      if account_data_temp.shape[0] <= c:
        account_data_temp['Target'] = 1
      else:
        account_data_temp_1 = account_data_temp.iloc[-c:]
        account_data_temp_1['Target'] = 1
        account_data_temp_2 = account_data_temp.iloc[:-c]
        account_data_temp = pd.concat([account_data_temp_2, account_data_temp_1])
    account_data_list.append(account_data_temp)   
  return pd.concat(account_data_list)

def TBV_difference(dataset):
  emp_ids = dataset.Emp_ID.unique()
  account_data_list = []
  for i in emp_ids:
    account_data_temp = dataset[dataset.Emp_ID == i]
    account_data_temp['TBV_difference'] = account_data_temp['Total Business Value'].shift(1)
    account_data_temp['TBV_difference'] = np.where(account_data_temp['TBV_difference'].isnull(),
                                                   account_data_temp['Total Business Value'],
                                                   account_data_temp['TBV_difference'])
    account_data_temp['TBV_difference'] = account_data_temp['Total Business Value'] - account_data_temp['TBV_difference']
    account_data_list.append(account_data_temp)
  return pd.concat(account_data_list)  


data = churn(train)
data = TBV_difference(data)
data['number_of_promotion'] = data['Designation'] - data['Joining Designation']
data["City"]= data["City"].astype('int64')
test_d = data[(data['MMM-YY']=='2017-12-01') & (data.Target==0)]
train_d = data[-((data['MMM-YY']=='2017-12-01') & (data.Target==0))]
train_d.drop('City',axis=1,inplace=True)
test_d.drop('City',axis=1,inplace=True)
from sklearn.ensemble import GradientBoostingClassifier
gsearch = GradientBoostingClassifier(random_state=321,verbose=0, max_depth=4).fit(train_d.drop(['MMM-YY','Emp_ID','Target'],axis=1), train_d['Target'])
test_d['Target']= gsearch.predict(test_d.drop(['MMM-YY','Emp_ID','Target'],axis=1))
test.merge(test_d[['Emp_ID','Target']],on='Emp_ID',how='left').to_csv('solution12.csv',index=False)
```
